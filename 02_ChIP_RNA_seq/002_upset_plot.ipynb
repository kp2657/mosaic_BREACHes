```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from upsetplot import UpSet
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
plt.rcParams['font.family'] = 'Helvetica'
```

```python
data = pd.read_csv('upset_plot_data.txt', sep='\t')
data.rename(
    columns={
        "135_Domains": "FXS_421",
        "FXS_Domains": "FXS_426",
        "GMO7730_Domains": "FXS_470",
        "GM09497_Domains": "FXS_448",        
    },
    inplace=True
)
data.set_index(list(data.columns), inplace=True)
```

```python
upset = UpSet(data, intersection_plot_elements=10)
fig = plt.figure(figsize=(10, 6))
upset.plot(fig=fig)

# Customize the plot
ax = fig.axes

for axis in ax:
    axis.grid(False)  # Remove gridlines
    if axis.get_ylabel():  # Customize y-axis labels if they exist
        axis.set_ylabel("")
    if axis.get_yticks().size > 0:  # Customize y-axis ticks
        axis.set_yticks(range(0, 13, 4))  # Example: Adjust y-axis ticks from 0 to 20 in steps of 5
        axis.set_yticklabels([])
        
        
plt.tight_layout()
#plt.savefig('NPC_upsetplot_top.pdf', dpi=1000)
plt.show()
```
